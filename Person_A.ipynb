```python
import pandas as pd
file_path = "VF_Hackathon_Dataset_India_Large.xlsx"
pdf = pd.read_excel(file_path)
column_rules = {
    'name': ("", 'str'),
    'phone_numbers': ("", 'str'), 
    'officialPhone': ("", 'str'),
    'email': ("", 'str'),
    'websites': ("", 'str'),
    'officialWebsite': ("", 'str'),
    'yearEstablished': (0, 'int64'),
    'facebookLink': ("", 'str'),
    'twitterLink': ("", 'str'),
    'linkedinLink': ("", 'str'),
    'instagramLink': ("", 'str'),
    'address_line1': ("", 'str'),
    'address_line2': ("", 'str'),
    'address_line3': ("", 'str'),
    'address_city': ("", 'str'),
    'address_stateOrRegion': ("", 'str'),
    'address_zipOrPostcode': ("", 'str'),
    'address_country': ("", 'str'),
    'address_countryCode': ("", 'str'),
    'facilityTypeId': ("", 'str'),
    'operatorTypeId': ("", 'str'),
    'affiliationTypeIds': ("", 'str'),
    'description': ('Unknown', 'str'),
    'numberDoctors': (0, 'int64'),
    'capacity': (0, 'int64'),
    'specialties': ('[]','str'),
    'procedure': ('Unknown', 'str'),
    'equipment': ('Unknown', 'str'),
    'capability': ('Unknown', 'str'),
    'recency_of_page_update': ("", 'str'),
    'distinct_social_media_presence_count': (0, 'int64'),
    'affiliated_staff_presence': ("", 'str'),
    'custom_logo_presence': ("", 'str'),
    'number_of_facts_about_the_organization': (0, 'int64'),
    'post_metrics_most_recent_social_media_post_date': ("", 'str'),
    'post_metrics_post_count': (0, 'int64'),
    'engagement_metrics_n_followers': (0, 'int64'),
    'engagement_metrics_n_likes': (0, 'int64'),
    'engagement_metrics_n_engagements': (0, 'int64'),
    'latitude': (0.0, 'float64'),
    'longitude': (0.0, 'float64')
}
print(column_rules.items())
for col, (default_val, dtype) in column_rules.items():
    if col in pdf.columns:
        # Fill nulls with your defined default
        pdf[col] = pdf[col].fillna(default_val)
        # Cast to the specific dtype
        pdf[col] = pdf[col].astype(dtype)
    else:
        print("lol"+str(col))
#for col in pdf.columns:
 #  pdf[col] = pdf[col].astype(str).replace('nan', None)
df = spark.createDataFrame(pdf)

df.write.format("delta").mode("overwrite").saveAsTable("Base_table")
```
